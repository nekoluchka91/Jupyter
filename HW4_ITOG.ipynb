## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling:
дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
```

Вывод: На графике отчетлива видна прямая линейная зависимость. Жилая площадь влияет на стоимость недвижимости. Чем больше жилая площадь дома, тем выше стоимость дома.

```python
plt.figure(figsize=(6, 4))

plt.scatter(df['price'], df['grade'])

plt.title('Зависимость стоимости от конструкции здания')
plt.xlabel('Стоимость')
plt.ylabel('Конструкция здания');
```

Вывод: График показывает зависимость стоимости недвижимости от конструкции здания. Чем лучше конструкция здания и дизайн, тем выше стоимость дома.
